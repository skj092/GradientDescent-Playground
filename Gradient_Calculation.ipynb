<h1>Gradient Calculation</h1>

<h2>Being a mathematics student I've always been curious to know how pytorch calculate derivative of different kind of functions. In this notebook I'll take different types of functions and explain how to calculate gradient or derivative using pytorch.<h2>

<h3>
Depending on the input and output dimensional we can classify function into four types. 
<ul>
<li>A function who take one dimensional input and produce one dimensional output.
<li>A function who take one dimensional input and produce multi dimensional output.
<li>A function who take multi dimensional input and produce one dimensional output.
<li>A function who take multi dimensional input and produce multi dimensional output.
<p>Mathematically these fumctions can be represented as follows: <p>
</ul><h3>

<div class = 'math'>
\begin{align}
    &f\colon R  \longrightarrow R\\
     &f\colon R  \longrightarrow R^n\\
      &f\colon R^n \longrightarrow R\\
       &f\colon R^n \longrightarrow R^n\\
\end{align}
<h3>Here R represent real number and m and n are integers with value greater then 1. </h3>
</div>

<h1>Lets Try to calculate derivative of these function.<h1>
<h2>1. Function with one dimensional input and one dimensional output.</h2>

```python
# importing the libraries
# import torch
# from torch.autograd import Variable
```

```python
# x = torch.tensor([2.], requires_grad=True)
# f = x**4
# df = f.backward()
# x.grad
```

<h3>We can also define x using Variable function and perform same operation as below. </h3>

```python
# x = Variable(torch.tensor([2.]), requires_grad=True)
# fx = x**4
# d_f = fx.backward(gradient=torch.ones(1))
# x.grad
```

<h3>Here we defined a function<h3>
<div class="math">
\begin{align*}
&f\colon R \longrightarrow R\\
&\ni \forall x \in R\\
    &f = x^4\\
     &\therefore df = f.backward() = \frac{df}{dx} = 4*x^3\\
      &\therefore x.grad = \frac{df}{dx}_{x=2} =4*2^3 = 32
\end{align*}
</div>

<h2>2. Function with multi dimensional input and one dimensional output.</h2>

```python
# x = Variable(torch.tensor([5.,4.]), requires_grad=True)
# f = (x**2).sum()
# df = f.backward()
# x.grad
```

<h3>Here </h3>
<div class = 'math'>
\begin{align*}
    &f\colon R^2 \longrightarrow R\\
     &\ni \forall  x = \left(x_{1},x_{2}\right) \in R^2\\
      &f = x_{1}^2+x_{2}^2\\
       &\therefore df = f.backward() = \frac{df}{dx} = \left(\frac {df}{dx_1},\frac{df}{dx_2}\right)\\
         &\because\frac{df}{dx_1} = 2*x_1 \\
           &\frac{df}{dx_2} = 2*x_2\\
             &\therefore{df} = \left(2x_1,2x_2\right)\\
               &\therefore x.grad = \frac{df}{dx}_{x=\left(5,4\right)} = \left(2*5,2*4\right) = (10,8)  
\end{align*}
</div>

<h2>3. Function with one dimensional input and multi dimensional output.</h2>
<h3>These types of function also called as vector valued function.</h3>

```python
# x = Variable(torch.tensor(4.), requires_grad=True)
# f = Variable(torch.ones(2,1), requires_grad=False)
# f[0] = x**2
# f[1] = x**3
# df = f.backward(gradient=torch.ones(f.size()))
# x.grad
```

<h3>Here</h3>
<div class = 'math'>
\begin{align*}
    &f\colon R \longrightarrow R^2\\
     &\ni \forall  x \in R\\
      &f(x) = (x^2,x^3) = x^2i+x^3j\\
       &df = f.backward() = \frac{df}{dx} = 2xi+3x^2j\\
        &\therefore x.grad =  \frac{df}{dx}_{x=4} = (2*4)i+(3*4^2)j = 8+48 = 56
\end{align*}
</div>

<h2>4. Function with multi dimensional input and multi dimensional output.</h2>

```python
# x = Variable(torch.tensor([2.,2.]), requires_grad=True)
# f = Variable(torch.ones(3,1), requires_grad=False)
# f[0] = x[0]**2
# f[1] = x[1]**3
# f[2] = x[1]**4
# df = f.backward(gradient=torch.ones(f.size()))
# x.grad
```

<div class = 'math'>
\begin{align*}
    &f\colon R^2 \longrightarrow R^3\\
     &\ni \forall  x = \left(x_{1},x_{2}\right) \in R^2\\
      &f(x) = \left(x_1^2,x_2^3,x_2^4\right)\\
       &df = f.backward() = \frac{df}{dx} = \left(\frac {df}{dx_1},\frac{df}{dx_2}\right)\\
        &\because\frac{df}{dx_1} = 2x_1 \\
         &\frac{df}{dx_2} = 3x_2^2+4x_2^3\\
          &\therefore df = \left(2x_1,3x_2^2+4x_2^3\right)\\
           &\therefore x.grad =  \frac{df}{dx}_{x=(2,2)} = (2*2,3*2^2+4*2^3) = (4,44)
\end{align*}  
</div>
